### 6.2. Create reusable functions  for plotting Categorical and Numerical values

```python
# Find if the column is categorical or numerical 

def check_data_type(data, col):
    if data[col].dtype == np.int64 or data[col].dtype == np.float64:
        return "numerical"
    else:
        return "categorical"
```

```python
# Univariate - function to analyse one single variable with respect to a target variable

def univariate(data, col, target_col, ylog=False, x_label_angle=False, h_layout=True):
    if check_data_type(data, col) == "numerical":
        sns.distplot(data[col], hist=False)

    elif check_data_type(data, col) == "categorical":
        val_count = data[col].value_counts()
        df1 = pd.DataFrame({col: val_count.index,'count': val_count.values})
        
        target_1_percentage = data[[col, target_col]].groupby([col],as_index=False).mean()
        target_1_percentage[target_col] = target_1_percentage[target_col]*100
        target_1_percentage.sort_values(by=target_col,inplace = True)
        
        # If the plot is not readable, use the log scale
        if(h_layout):
            fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(15,7))
        else:
            fig, (ax1, ax2) = plt.subplots(nrows=2, figsize=(25,35))
            
            
        # 1. Subplot 1: Count plot of the column
        s = sns.countplot(ax=ax1, x=col, data=data, hue=target_col)
        ax1.set_title(col, fontsize = 20)
        ax1.legend(['Loan Repayer','Loan Defaulter'])
        ax1.set_xlabel(col,fontdict={'fontsize' : 15, 'fontweight' : 3})
        
        if(x_label_angle):
            s.set_xticklabels(s.get_xticklabels(),rotation=75)
            
        # 2. Subplot 2: Percentage of defaulters within the column
        s = sns.barplot(ax=ax2, x = col, y=target_col, data=target_1_percentage)
        ax2.set_title("Defaulters perrcentage in "+col, fontsize = 20)    
        ax2.set_xlabel(col,fontdict={'fontsize' : 15, 'fontweight' : 3})
        ax2.set_ylabel(target_col,fontdict={'fontsize' : 15, 'fontweight' : 3})
        
        if(x_label_angle):
            s.set_xticklabels(s.get_xticklabels(),rotation=75)
            
        
        # If the plot is not readable, use the log scale          
        if ylog:
            ax1.set_yscale('log')
            ax1.set_ylabel("Count (log)",fontdict={'fontsize' : 15, 'fontweight' : 3})
        else:
            ax1.set_ylabel("Count",fontdict={'fontsize' : 15, 'fontweight' : 3})

        
        plt.show()
```

```python
# Plot the repetitive rel plots in bivaritae numerical analysis

def bivariate_n(x, y, data, hue, kind, labels):

    plt.figure(figsize=[15,15])
    sns.relplot(x=x, y=y, data=data, hue=hue, kind=kind,legend = False)
    plt.legend(labels=labels)
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

```python
# Plot the repetitive barplots in bivariate categorical analysis

def bivariate_c(x, y, data, hue, figsize, labels, title):
    
    plt.figure(figsize=figsize)
    sns.barplot(x=x, y=y, data=data, hue=hue)
    
    plt.xlabel(x,fontsize = 15)    
    plt.ylabel(y,fontsize = 15)    
    plt.title(title,fontsize = 20) 
    plt.xticks(rotation=45, ha='right')
    plt.legend(labels = labels )
    plt.show()
```

```python
# Plot the repetitive countplots in univariate categorical analysis on the merged dataframe

def univariate_c_merged(col, data, hue, palette, ylog, figsize):
    plt.figure(figsize=figsize)
    ax=sns.countplot(x=col, data=data, hue=hue, palette=palette, order=data[col].value_counts().index)

    if ylog:
        plt.yscale('log')
        plt.ylabel("Count (log)", fontsize=15)     
    else:
        plt.ylabel("Count", fontsize=15)       

    plt.title(col , fontsize=20) 
    plt.legend(loc = "upper right")
    plt.xticks(rotation=45, ha='right')
    
    plt.show()
```

```python
# Function to plot point plots 

def pointplot(data,hue,x,y):
    plt.figure(figsize=(12,6))
    sns.pointplot(x=x, y=y, hue=hue, data=data)
    plt.title(x+" vs "+y, fontsize = 15)
```
